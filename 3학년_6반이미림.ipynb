```python
import matplotlib
import matplotlib.pyplot as plt
matplotlib.rcParams['font.family']='NanumGothic'  #나눔고딕 세팅
matplotlib.rcParams['font.size']=15 #글자 크기
matplotlib.rcParams['axes.unicode_minus']=False #마이너스 깨짐 방지
```

```python
#엑셀파일 불러오기
import pandas as pd
df=pd.read_excel('/content/인천광역시_·_산업_·_종사자지위별__종사자수_06____20230530190719.xlsx')
df
```

```python
plt.plot(df['2010'],df['산업별'])
plt.plot(df['2011'],df['산업별'])
plt.plot(df['2012'],df['산업별'])
plt.plot(df['2013'],df['산업별'])
plt.plot(df['2014'],df['산업별'])
plt.plot(df['2015'],df['산업별'])
plt.plot(df['2016'],df['산업별'])
plt.legend(['2010','2011', '2012','2013','2014','2015','2016'])
plt.savefig('인천_산업별_종사자수.png')
```
