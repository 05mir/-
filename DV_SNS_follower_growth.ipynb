```python
import csv

data = [['Month', 'Instagram', 'Twitter', 'Facebook', 'LinkedIn'],
        ['Jan', 300, 200, 150, 100],
        ['Feb', 350, 250, 180, 110],
        ['Mar', 400, 280, 200, 130],
        ['Apr', 450, 320, 250, 150],
        ['May', 500, 350, 280, 170],
        ['Jun', 550, 380, 300, 190],
        ['Jul', 600, 420, 350, 210],
        ['Aug', 650, 460, 400, 230],
        ['Sep', 700, 500, 450, 250],
        ['Oct', 750, 550, 500, 280],
        ['Nov', 800, 600, 550, 310],
        ['Dec', 850, 650, 600, 340]]

with open('social_media_growth.csv', mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerows(data)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# csv 데이터 가져오기
data = pd.read_csv("social_media_growth.csv")

#각 플랫폼에서 데이터 추출
dates = data['Month']
instagram = data['Instagram']
twitter = data['Twitter']
facebook = data['Facebook']

# 그래프 생성
fig, ax = plt.subplots()
ax.plot(dates, instagram, label='Instagram',marker='o',markersize=10)
ax.plot(dates, twitter, label='Twitter',marker='s',markersize=10)
ax.plot(dates, facebook, label='Facebook',marker='^',markersize=10)
ax.set_xlabel('Date')
ax.set_ylabel('Number of Followers')
ax.set_title('Social Media Follower Growth')
ax.legend()

plt.show()
```
